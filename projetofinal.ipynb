```python
#Construa uma tabela auxiliar que sumarize o valor vendido por cada vendedor, ordenando do maior para o menor;
import csv
import pandas as pd
import re

df = pd.read_csv("DB_Teste.csv", sep=(";"))

sales_by_seller = df.groupby('Vendedor')['Valor'].sum()

vendas_por_vendedor = str(sales_by_seller).split('Vendedor ')

venda_separada = []
cont = 0

for venda in vendas_por_vendedor:
  if cont != 0:
    venda_separada.append(str(venda).split())
  cont += 1

ordem_vendedores = []

for i in venda_separada:
  novo_i = str(i).split(" ")
  vendedor = novo_i[0]
  if vendedor == ['Vendedor']:
    pass
  else:
    if vendedor[3] not in '0123456789':
      ordem_vendedores.append(vendedor[2])
    else:
      ordem_vendedores.append(vendedor[2] + vendedor[3])


print("\nTabela que sumarize o valor vendido por cada vendedor, ordenando do maior para o menor: ")
lista_de_somas = []
for x in range(len(ordem_vendedores)):

    numeros = re.findall(r'\d+', sales_by_seller[x])
    numeros_float = list(map(float, numeros))

    soma = sum(numeros_float)

    lista_de_somas.append(soma)

nova_lista = list(zip(ordem_vendedores, lista_de_somas))
nova_lista.sort(key=lambda x: x[1], reverse=True)

for i, j in nova_lista:
  print(f'Vendedor {i} = R$ {j:.2f}')
```

```python
#Imprima e identifica qual foi o cliente responsável pela venda com maior valor e com menor valor;
import csv
import pandas as pd
import re

def processa_vendas(arquivo_csv):
    # Lendo o arquivo CSV
    df = pd.read_csv("DB_Teste.csv", sep=(";"))

    # Substituindo 'R$' pela vírgula no valor das vendas
    df['Valor'] = df['Valor'].str.replace('R$', '')

    # Converter coluna de 'Valor' para float
    df['Valor'] = df['Valor'].astype(float)

    # Encontrando o cliente responsável pela venda com maior valor
    maior_valor = df['Valor'].max()
    maior_cliente = df[df['Valor'] == maior_valor]['Cliente'].values[0]

    # Encontrando o cliente responsável pela venda com menor valor
    menor_valor = df['Valor'].min()
    menor_cliente = df[df['Valor'] == menor_valor]['Cliente'].values[0]

    # Retornando os resultados
    return maior_cliente, maior_valor, menor_cliente, menor_valor


# Entrada do usuário: o arquivo CSV
arquivo_csv = 'DB_Teste.csv'

# Processando as vendas e obtendo os resultados
maior_cliente, maior_valor, menor_cliente, menor_valor = processa_vendas(arquivo_csv)

# Imprimindo os resultados
print("O cliente responsável pela venda com maior valor foi {} e o valor foi R${:.2f}.".format(maior_cliente, maior_valor))
print("O cliente responsável pela venda com menor valor foi {} e o valor foi R${:.2f}.".format(menor_cliente, menor_valor))
```

```python
#Imprima valor médio por Tipo de venda (Serviços, Licenciamento, Produtos)
import pandas as pd
import re

def processa_vendas(arquivo_csv):
    # Lendo o arquivo CSV
    df = pd.read_csv("DB_Teste.csv", sep=(";"))

    # Substituindo 'R$' pela vírgula no valor das vendas
    df['Valor'] = df['Valor'].str.replace('R$', '')

    # Converter coluna de 'Valor' para float
    df['Valor'] = df['Valor'].astype(float)


    # Calculando o valor médio por tipo de venda
    media_por_tipo = df.groupby('Tipo de Venda')['Valor'].mean()

    # Retornando os resultados
    return media_por_tipo


# Entrada do usuário: o arquivo CSV
arquivo_csv = 'DB_Teste.csv'

# Processando as vendas e obtendo os resultados
media_por_tipo = processa_vendas(arquivo_csv)

# Imprimindo os resultados

print("\nValor médio por tipo de venda:")
for tipo, valor in media_por_tipo.items():
    print("- Tipo: {} | Valor médio: R${:.2f}".format(tipo, valor))
```

```python
#Imprima o número de vendas realizada por cliente;
import pandas as pd

def quantidade_compras(arquivo_csv):
    # Lê o arquivo CSV e armazena na variável 'df'
    df = pd.read_csv("DB_Teste.csv", sep=(";"))

    # Cria um dicionário para armazenar a quantidade de compras de cada cliente
    quantidade_compras = {}

    # Itera sobre cada linha do DataFrame
    for _, row in df.iterrows():
        # Se o cliente já está no dicionário, incrementa a quantidade de compras
        if row['Cliente'] in quantidade_compras:
            quantidade_compras[row['Cliente']] += 1
        # Caso contrário, adiciona o cliente ao dicionário com a quantidade de compras igual a 1
        else:
            quantidade_compras[row['Cliente']] = 1

    # Imprime a quantidade de compras realizadas por cliente
    for cliente, quantidade in quantidade_compras.items():
        print(f"{cliente}, Quantidade de compras: {quantidade}")

# Chama a função para processar o arquivo CSV fornecido
quantidade_compras('clientes_vendas.csv')
```
